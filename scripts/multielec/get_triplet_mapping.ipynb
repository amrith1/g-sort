```python
import src.electrode_map as emap
import numpy as np
from scipy.io import savemat
import matplotlib.pyplot as plt
```

```python
triplets = []
for e, neighbors in enumerate(emap.LITKE_519_ARRAY_ADJ_MAT):
#     print(e)
    for n1 in neighbors:
        for n2 in neighbors:
            if (n1 != e) and (n2 != e) and (n1 != n2):
                if n2 in emap.LITKE_519_ARRAY_ADJ_MAT[n1]:
                    triplets += [sorted((e, n1, n2))]
triplets = np.array(triplets)
triplets_519 = np.unique(triplets, axis = 0 )
```

```python
triplets_519
```

```python
coords = emap.get_litke_array_coordinates_by_array_id(1502)
plt.figure()
fig, ax = plt.subplots(1,1,figsize=(7, 7))
ax.plot(coords[:,0], coords[:,1], '.', color = 'black')
j = 0
ax.plot(coords[triplets_519[j],0], coords[triplets_519[j],1], '.', color = 'red')
```
