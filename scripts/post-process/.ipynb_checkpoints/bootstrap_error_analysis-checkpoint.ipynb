```python
def get_instance_modified_edge_probs(edge_probs, error, dss_inds, average_error, std_error, factor, test_set):
    good_dss_inds = []
    test_set = np.array(test_set)
    for i1, i2, i3 in zip(error, edge_probs, dss_inds):
        i4 = []
        for j1, j2, j3 in zip(i1,i2, i3):
            j4 = []
            for k1, k3 in zip(j1, j3):
                k4 = []
                if np.all(k1[test_set] < average_error[test_set] -factor * std_error[test_set]) :
                    
                    k4 += [j3]
                j4 += [k4]
            i4 += [j4]
        good_dss_inds += [i4]
```
